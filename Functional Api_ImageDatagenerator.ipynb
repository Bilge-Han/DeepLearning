```python
# Hafta 11
#-Functional Api :
#-ImageDatagenerator :  Veri okuma, çoğaltma vsvsvsvsvs...
#   - Veriyi okuyabilmek için, verinin hazır bulunurluk seviyesi olması lazım. belli formatta olması lazım
#   - Eğitimde çoğaltma imkanı
#   - Klasörden veriyi okuduğu için kendisi otomatik etiketliyo, biz uğraşmıyoz
#   - İstediğimiz batchlar halinde veriyi getiriyo...
#
#   - Görüntülerin ilgili etiketle klasörlerde ayrılmış olması lazım
#   - Test ve train ve validation olarak klasörlere ayırmamız lazım
# Return
# Yield : bütün veriyi hemen yüklemek yerine ihtiyacı olduğu kadarını-zamanda rame yüklüyor okuyor... - daha etkin hafıza kullanıyoruz.
```

```python
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, AveragePooling2D, MaxPool2D
from keras.datasets import mnist
from keras.utils import to_categorical
```

```python
# ImageDataGenerator ile Data Augmentation (Veri Çoğaltma)
from numpy import expand_dims
import matplotlib.pyplot as plt

from keras.preprocessing.image import ImageDataGenerator
from keras.utils import load_img,img_to_array


img=load_img('./drive/MyDrive/Deep Learning/spos.png')

data=img_to_array(img)
data.shape

# Örnek atalım bir tane
samples=expand_dims(data,0)
samples.shape

# Data gen tanımlayalım
dataGen=ImageDataGenerator(
    rotation_range= 45, #45dereclik açılarla döndür
    vertical_flip=True, #dikey çevir
    horizontal_flip=True,
    zoom_range=0.7
)

# Yeni görüntüler oluşturalım
it=dataGen.flow(samples,batch_size=1)

plt.figure(figsize=(10,5))

for i in range(128):
  batch=it.next()
  plt.subplot(8,16,i+1)
  print(batch.shape)
  image=batch[0].astype('uint8')
  plt.imshow(image)
  plt.axis('off')
plt.show()
```

```python
# ImageDataGenerator ile Veri Okuma

train_path='./drive/MyDrive/Deep Learning/chest_xray/train'
test_path='./drive/MyDrive/Deep Learning/chest_xray/test'

trainGen=ImageDataGenerator(
    rescale=1./255,
    rotation_range=45,
    vertical_flip=True,
    horizontal_flip=True,
    zoom_range=0.5,
    validation_split=0.1      # validation verisi ayırdı.
)

valGen=ImageDataGenerator(rescale=1./255)
testGen=ImageDataGenerator(rescale=1./255)

# Veri okuma yöntemleri
# flow : görüntüleri okursunuz, listeye atarsanız, liste içerisinden flow(okuyabilirsiniz) atabilirsiniz
# flow_from_directory : bir directory'den oku listeye atma, ihtiyacım olduğunda getir

trainData=trainGen.flow_from_directory(
    train_path,               # bu directory'den oku
    target_size=(224,224),    # hedef ne olsun, tüm görüntüler 28'e 28 olsun dedik, ne kadar büyük okursak o kadar kaliteli görüntü çıkar.
    class_mode= 'binary',     # okuma türünü belirliyoruz  -categorical dersek= 2D one-hut encoding yapılmış şekilde okur, -binary dersek= binary şekilde bir bir okur
    batch_size=8,
    subset='training'
    )

valData=trainGen.flow_from_directory(
    test_path,
    target_size=(224,224),
    class_mode= 'binary',
    batch_size=8,
    subset='validation'      # ayırdığımız validation datayı kullandık.
    )

testData=trainGen.flow_from_directory(
    test_path,
    target_size=(224,224),
    class_mode= 'binary',
    batch_size=8,
    )

print(testData.class_indices)  # hangi etikete hangi indisi attı
print(testData.classes)        # classları görürüz
print(testData.class_mode)     # class mode görürüz
print(testData.filenames)


plt.figure(figsize=(10,10))

for i in range(6):
  img,y=testData.next()
  plt.subplot(2,3,i+1)
  print(img.shape)
  plt.imshow(img[0])
  plt.title(y[0])
  plt.axis('off')
plt.show()
```

```python
# LeNet-5 Functional Api modelini eğitelim bunun ile

# Kullanmaya alışalım bu önemli...

from keras.layers import Input
from keras.models import Model

input = Input(shape=(224,224,3)) # girdi

x=Conv2D(6,kernel_size=(5,5),activation='tanh',strides=(1,1))(input)  # girdiler köşeden alınır o yüzden girdiyi vermemiz lazım
x=AveragePooling2D(strides=(2,2))(x) # burda da bir önceki katmanı veriyoruz(x)

x=Conv2D(16,kernel_size=(5,5),activation='tanh',strides=(1,1))(x)
x=AveragePooling2D(strides=(2,2))(x)

x=Flatten()(x)

x=Dense(120,activation='tanh')(x)
x=Dense(84,activation='tanh')(x)
output=Dense(1,activation='sigmoid')(x)

#modeli oluşturabiliriz
model = Model(inputs=input,outputs=output)

model.summary()

# Modeli derleyelim eğitelim
model.compile(loss="binary_crossentropy",optimizer="sgd",metrics=['accuracy'])

history = model.fit(trainData,epochs=15,validation_data=valData)
```

```python
from keras.utils.vis_utils import plot_model
plot_model(model,to_file='model_plot.png',show_shapes=True,show_layer_names=True)
```

```python
import matplotlib.pyplot as plt

print(history.history.keys())
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'valid'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'valid'], loc='upper left')
plt.show()
```
